```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
from PIL import Image # converting images into arrays

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0

# install wordcloud
#!conda install -c conda-forge wordcloud==1.4.1 --yes
#!conda install -c conda-forge wikipedia --yes

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
import wikipedia as wp

from bs4 import BeautifulSoup

import requests
print ('Wordcloud is installed and imported!')
```

```python
html = wp.page("Cuisine_of_New_York_City").html().encode("UTF-8")
#html
```

```python
soup = BeautifulSoup(html, 'lxml')

#print(soup.prettify())
data = ''
```

```python
#article = soup.find_all('div',class_ = "div-col columns column-width")
article = soup.find_all('div', class_ = "div-col columns column-width")

flag = 0
for item in article:
    data = data + item.get_text()
    
record = data.split('\n')[1:56]
#record[0].split(' ')
record
```

```python

file = open("NYCuisine.txt","w") 


for every_record in record:
    
    every_cuisine = every_record.split(' ')
    
    for every_word in every_cuisine:
        if every_word != '-' or every_word != '(' or every_word != ')' or every_word.isdigit() == False:
            file.write(every_word) 
 
file.close() 
```

```python
cuisine = open('NYCuisine.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
NYCloud = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
NYCloud.generate(cuisine)
```

```python
# display the word cloud
plt.imshow(NYCloud, interpolation='bilinear')
plt.axis('off')

fig = plt.figure()
fig.set_figwidth(30)
fig.set_figheight(45)

plt.show()
```
